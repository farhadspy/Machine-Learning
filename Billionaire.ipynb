```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Farhad/Desktop/python/Machine_Learning/Billionaire.csv", encoding='latin1')
data.head(10)
```

preprocessing

قبل از اینکه کار رو شروع کنبم ، ببینیم که مقادیر خالی یا نال داریم یا نه؟

```python
data.isnull().sum()
```

این 79 نفر رو حذف می کنیم

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

delete B and $ from networth column

```python
data["NetWorth"] = data["NetWorth"].str.strip("$")
data["NetWorth"] = data["NetWorth"].str.strip("B")
data["NetWorth"] = data["NetWorth"].astype(float)
data.head(10)
```

process

تغییر ترتیب از کمترین پولدار به بیشترین پولدار
ascending=True

```python
df = data.sort_values(by="NetWorth", ascending=False).head(10).reset_index()
df
```

```python
plt.figure(figsize=(24, 8))
plt.bar(df["Name"], df["NetWorth"])
plt.show()
```

```python
df = data["Source"].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df["count"], labels=df["Source"])
plt.title("Top 10 Domains to become a Billionaire")
plt.show()
```

```python
df = data["Industry"].value_counts().reset_index()
df
```

```python
plt.figure(figsize=(24, 10))
plt.pie(df["count"], labels=df["Industry"])
plt.title("Top 10 Industry to become a Billionaire")
plt.show()
```

```python
df = data["Country"].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df["count"], labels=df["Country"])
plt.title("Top 5 Countries to become a Billionaire")
plt.show()
```
